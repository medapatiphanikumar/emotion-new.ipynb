```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
tf.__version__
```

```python
# adjust the vocabulary size 
vocab_size=5000
```

```python
# the most important steps for any type of lstm neural network:
# 1.one hot representation- to give proper index postion to the words in sentence
# 2. embedding layer using keras to genrate embedding matrix
# 3. adjust the vocab_size =10000(limit),dimensions = n n is based on the sentence lenght
# dimensions are nothing but the number of elements in the vector
```

```python
mess = X.copy()
```

```python
mess.reset_index(inplace=True) # to replace and restore the nan value space
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(mess)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', mess['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus # in order to remove the stopwords
```

```python
onehot_repr=[one_hot(words,vocab_size)for words in corpus] 
onehot_repr
```

```python
sen_len = 20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sen_len)
print(embedded_docs)
```

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(vocab_size,embedding_vector_features,input_length=sen_len))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
X_new = np.array(embedded_docs)
y_new = np.array(y)
```

```python
X_new.shape
```

```python
y_new.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_new,y_new,test_size=0.2)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
from sklearn.metrics import accuracy_score
```

```python
z_lstm = model.predict(X_test)
```

```python
print(accuracy_score(y_test,z_lstm.round())*100)
```
